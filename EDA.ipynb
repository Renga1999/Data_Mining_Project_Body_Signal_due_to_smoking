```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
data = pd.read_csv('smoking.csv')
data.head()
```

```python
ax = sns.scatterplot(x='height(cm)', y='hemoglobin',hue="smoking", data=data)
plt.xlabel("Height")
plt.ylabel("Hemoglobin")
plt.title("Height vs Hemoglobin")
plt.show()
```

```python
sns.catplot(data=data, x="height(cm)", y="weight(kg)", kind="box")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```

```python
ax = sns.scatterplot(x='AST', y='ALT',hue="smoking", data=data)
plt.xlabel("AST")
plt.ylabel("ALT")
plt.title("AST vs ALT")
plt.show()
```

```python
ax = sns.boxplot(x='gender', y='height(cm)', data=data, color='#ADD8E6')
plt.xlabel("Gender")
plt.ylabel("Height")
plt.title("Gender vs Height")
plt.show()
```

```python
ax = sns.scatterplot(x='Cholesterol', y='LDL',hue="smoking", data=data)
plt.xlabel("Cholestrol")
plt.ylabel("LDL")
plt.title("Cholesterol vs LDL")
plt.show()
```

```python
ax = sns.boxplot(x='smoking', y='hemoglobin', data=data, color='#CBC3E3')
plt.xlabel("Smoking")
plt.ylabel("Hemoglobin")
plt.title("Smoking vs Hemoglobin")
plt.show()
```

```python
ax = sns.boxplot(x='smoking', y='height(cm)', data=data, color='#99c2a2')
plt.xlabel("Smoking")
plt.ylabel("Height")
plt.title("Smoking vs Height")
plt.show()
```

```python
ax = sns.scatterplot(x='weight(kg)', y='waist(cm)',hue="smoking", data=data)
plt.xlabel("Cholestrol")
plt.ylabel("LDL")
plt.title("Cholesterol vs LDL")
plt.show()
```

```python
ax = sns.boxplot(x='systolic', y='relaxation',hue="smoking", data=data)
plt.xlabel("Cholestrol")
plt.ylabel("LDL")
plt.title("Cholesterol vs LDL")
plt.show()
```

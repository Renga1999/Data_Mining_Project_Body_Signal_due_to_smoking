```python
#ANOVA test on ALT vs Smoking
model = ols('ALT ~ smoking', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
#ANOVA test on GTP vs Smoking
model = ols('GTP ~ smoking', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
#ANOVA test on Oral vs Smoking
model = ols('oral ~ smoking', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
#ANOVA test on Dental Caries vs Smoking
model = ols('dentalcaries ~ smoking', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
#ANOVA test on Tartar vs Smoking
model = ols('tartar ~ smoking', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

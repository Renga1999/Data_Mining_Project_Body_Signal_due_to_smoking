```python


from imblearn.over_sampling import SMOTENC
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix


```

```python

X = df.iloc[:,0:26]
X
y= df["smoking"]
y = pd.DataFrame(y,columns = ["smoking"])
y


```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
y_train.value_counts()
```

```python
train = pd.concat([X_train,y_train],axis=1)
test = pd.concat([X_test,y_test],axis=1)
train
X.dtypes
```

```python
# balancing the data out
sm = SMOTENC(random_state=42,categorical_features=[1,23,25])
X_res, y_res = sm.fit_resample(X_train, y_train)
```

```python
# percentage of values after balancing
y_res.value_counts(normalize = True)
```

```python
X_res.columns
```

```python
# getting dummies for train 
X_res = pd.get_dummies(X_res)
X_res
X_res = X_res.iloc[:,0:26]
```

```python
# getting dummies for test
X_test = pd.get_dummies(X_test)
X_test = X_test.iloc[:,0:26]
```

```python
X_train = X_res
y_train = y_res
```

MODELING - Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
features = X_train.columns
features
```

```python
dtree = DecisionTreeClassifier(random_state=42)
dtree1 = dtree.fit(X_train, y_train)
```

```python
clf1 = DecisionTreeClassifier( random_state = 42)
model2  = clf1.fit(X_train, y_train)


tree.plot_tree(clf1, feature_names=features, rounded = True, filled = True,fontsize=14)
plt.show()
```

```python
# prediction 
y_Pred = model2.predict(X_test)

#evaluation
print(accuracy_score(y_test, y_Pred))
print(confusion_matrix(y_test, y_Pred))
print(classification_report(y_test, y_Pred))
```
